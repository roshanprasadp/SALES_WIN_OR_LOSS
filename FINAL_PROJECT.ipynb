# STEP 4: FEATURE SELECTION: Feature Selection is one of the core concepts in machine learning which hugely impacts the performance of any model. The data features that you use to train your machine learning models have a huge influence on the performance you can achieve. Not all the features are important to train the model if we deal with continous data go with ANOVA, if we deal with categorical data go with CHISQUARE.

https://towardsdatascience.com/feature-selection-techniques-in-machine-learning-with-python-f24e7da3f36e   #for reference purpose

In this dataset there are some continous and categorical data.So, we have to apply both ANOVA and CHISQUARE testing to get the better features. ANOVA(Analysis of variance): if the data is continous and categorical or vice-versa we gotta go with this anova testing. If the value of p is less than significance(0.05) that indicates it is an important feature, If valuue of p is grater than the significance(0.05) that indicates it is not an important feature so we can remove the feature.

Chisquare testing: if the data is categorical we gotta go with this Chisquare testing. If the value of p is less than significance(0.05) that indicates it is an important feature, If valuue of p is grater than the significance(0.05) that indicates it is not an important feature so we can remove the feature.
